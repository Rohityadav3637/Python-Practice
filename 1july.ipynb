```python
# Seaborn: statistical data visualization library based on matplotlib
import seaborn as sns 

# Matplotlib: comprehensive library for creating static, animated, and interactive visualizations
import matplotlib.pyplot as plt 

# Numpy: fundamental package for scientific computing with Python
import numpy as np  

# Pandas: data analysis and manipulation tool
import pandas as pd 
```

```python
# Load the 'tips' dataset
# Tips dataset: records of restaurant bills and tips, used for demonstrating Seaborn plots

tips = sns.load_dataset('tips')
```

```python
# showing first 5 rows of dataset
tips.head()
```

```python
# Line Plot: used to visualize trends over time or continuous data
plt.figure(figsize=(10, 6))
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='day', markers=True, dashes=False)
plt.title('Line Plot of Tips by Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.grid(True)
plt.show()
```

```python
# Histogram: used to show the distribution of a single variable
# KDE Plot: Kernel Density Estimate, a smoothed version of the histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x='total_bill', kde=True, bins=30, color='blue')
plt.title('Histogram and KDE of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Bar Plot: used to show the relationship between a categorical variable and a continuous variable

plt.figure(figsize=(10, 6))
sns.barplot(data=tips, x='day', y='total_bill', ci='sd', palette='muted')
plt.title('Bar Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
# Heatmap: used to visualize data in matrix form, showing relationships between variables
# Correlation Heatmap: specifically used to show correlation coefficients between variables

# Select only the numeric columns from the dataset
numeric_tips = tips.select_dtypes(include=[np.number])

# Compute correlation matrix
corr = numeric_tips.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
# Scatterplot: used to visualize the relationship between two continuous variables

plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='time', palette='deep')
plt.title('Scatterplot of Tips by Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.grid(True)
plt.show()
```

```python
# KDE Plot: Kernel Density Estimate, a smoothed version of the histogram
# Rug Plot: small marks along the axis to show individual data points

plt.figure(figsize=(10, 6))
sns.kdeplot(data=tips, x='total_bill', shade=True, color='green')
sns.rugplot(data=tips, x='total_bill', color='black')
plt.title('KDE Plot of Total Bill with Rug Plot')  # Add title
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
# Box Plot: used to show the distribution of a dataset based on a five-number summary (minimum, first quartile, median, third quartile, and maximum)

plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker', palette='coolwarm')
plt.title('Box Plot of Total Bill by Day and Smoker')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
# Violin Plot: combines the features of a box plot and a KDE plot, showing distribution of the data and its probability density

plt.figure(figsize=(10, 6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True, palette='muted')
plt.title('Violin Plot of Total Bill by Day and Sex')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
# Pair Plot: creates a grid of scatter plots for each pair of variables in the dataset

sns.pairplot(tips, hue='sex', palette='coolwarm')
plt.show()
```

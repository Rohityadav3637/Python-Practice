```python
# Import necessary libraries

import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
# Sample data for the charts

x = np.linspace(0, 10, 100)
y = np.sin(x)
data = [10, 20, 30, 40, 50]
categories = ['A', 'B', 'C', 'D', 'E']
```

```python
# Line Chart

plt.figure(figsize=(10, 5))  
plt.plot(x, y, label='Sine Wave', color='blue', linestyle='-', linewidth=2)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Chart')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bar Chart

plt.figure(figsize=(10, 5))
plt.bar(categories, data, color='green')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.grid(True)
plt.show()
```

```python
# Pie Chart

plt.figure(figsize=(8, 8))
plt.pie(data, labels=categories, colors=['red', 'green', 'blue', 'yellow', 'purple'], 
        explode=[0, 0.1, 0, 0, 0], shadow=True, startangle=140, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# Histogram

plt.figure(figsize=(10, 5))
data_hist = np.random.randn(1000)
plt.hist(data_hist, bins=30, color='blue', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.grid(True)
plt.show()
```

```python
# Scatter Plot

plt.figure(figsize=(10, 5))
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
plt.scatter(x_scatter, y_scatter, color='red')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Scatter Plot')
plt.grid(True)
plt.show()
```

```python
# Area Chart

plt.figure(figsize=(10, 5))
plt.fill_between(x, y, color='orange', alpha=0.5)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Area Chart')
plt.grid(True)
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(10, 5))
data_box = [np.random.randn(100) for _ in range(5)]
plt.boxplot(data_box, patch_artist=True, 
            boxprops=dict(facecolor='cyan', color='blue'), 
            whiskerprops=dict(color='blue'),
            medianprops=dict(color='red'))
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Box Plot')
plt.grid(True)
plt.show()
```
